```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()#加载鸢尾花数据集
df=pd.DataFrame(iris.data)
print(df)
print(df.to_csv('a.csv',index=None,header=False))#存入csv文件中
```

```python
iris_data=iris.data#数据
feature_names=iris.feature_names#特征名
iris_target=iris.target#标签
print(feature_names)
print(iris_data.shape,iris_target.shape)
```

```python
import numpy as np
iris_target.shape=(150,1)
iris_all=np.hstack((iris_data,iris_target))#将数据和标签一一对应起来！！！！！！！
print('********************')
print(iris_all)
iris_data_df=pd.DataFrame(iris_data,columns=feature_names)
print('********************')
print(iris_data_df)
iris_target_df=pd.DataFrame(iris_target,columns=['target'])
print('********************')
print(iris_target_df)
iris_data_all_df=pd.DataFrame(iris_all,columns=feature_names+['target'])
print('********************')
print(iris_data_all_df)
print(iris_data_all_df.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(data=iris_data_all_df)#箱线图
plt.show()
```

```python
plt.plot(iris_data_df)
plt.legend(feature_names)#给出图中小方框里的内容
plt.show()
```

```python
sepal_data_df=iris_data_df[['sepal length (cm)','sepal width (cm)']]
plt.plot(sepal_data_df)
plt.legend(['sepal length (cm)','sepal width (cm)'])#可以把这句注释掉，对比一下看看
plt.title('sepal data')
plt.show()
```

```python
sns.pairplot(iris_data_all_df,vars=iris_data_all_df.columns[:4],hue='target',size=3,kind="reg")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#画协方差图
Corr_Mat=iris_data_df.corr()
Mat_img=plt.matshow(Corr_Mat,cmap=plt.cm.winter_r)
plt.colorbar(Mat_img,ticks=[-1,0,1])
plt.show()
#从这个图中可以看出，第二个特征（标号为1）与其他几个关系不大
```

```python
#降维
pca=PCA(n_components=2)
pca_2c=pca.fit_transform(iris_data_df)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
plt.scatter(pca_2c[:,0],pca_2c[:,1],alpha=0.8,cmap=plt.cm.winter)
plt.show()
```
