```python
filename="十三个原因.txt"
with open(filename) as f:
    text=f.read()
```

```python
import jieba
mytext=" ".join(jieba.cut(text))
print(mytext)
```

```python
def stopwordlist(filepath):
    stopwords=[line.strip() for line in open(filepath,'r',encoding='utf-8').readlines()]
    return stopwords
stopwordlist("stopwords.txt")
```

```python
def seg_sentence(sentence):
    sentence_seged=jieba.cut(sentence.strip())
    jieba.load_userdict("userdict.txt")#这句是加载用户自定义的词典，例：“走廊 上”就变成“走廊上”
    stopwords=stopwordlist('stopwords.txt')
    outstr=''
    for word in sentence_seged:
        if word not in stopwords:
            if word !='\t':
                outstr+=word
                outstr+=" "
    return outstr
f1=open("处理后.txt",'w')
with open("十三个原因.txt") as f:
    for line in f.readlines():
        seg_line=seg_sentence(line)
        f1.write(seg_line)
        print(seg_line)
```

```python
with open("处理后.txt") as f1:
    text=f1.read()
print(text)
```

```python
import jieba.analyse
keywords=jieba.analyse.extract_tags(text,topK=20,withWeight=True,allowPOS=())
for item in keywords:
    print(item[0],item[1])
```
