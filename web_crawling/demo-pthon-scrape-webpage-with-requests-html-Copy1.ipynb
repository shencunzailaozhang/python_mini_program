```python
from requests_html import HTMLSession
```

```python
session=HTMLSession()
```

```python
url='https://www.jianshu.com/p/85f4624485b9'
```

```python
r=session.get(url)#将链接对应的网页取到r中
```

```python
r.html.html
```

```python
print(r.html.text)#打印网页中的文字
```

```python
r.html.links#打印网页中所有链接（包括相对链接和绝对链接），
#相对地址：就是被链接文件相对于当前页面的地址。相也可当相应、类似相关的地址，如参考地址。
#绝对地址：就是文件在网络或本地的绝对位置。绝对，具有唯一性，如官网地址。
```

```python
r.html.absolute_links
```

```python
sel='body > div.note > div.post > div.article > div.show-content > div > p:nth-child(4) > a'#这是复制了网页源代码里的东西
```

```python
results=r.html.find(sel)
```

```python
def get_text_link_from_sel(sel):
    mylist=[]
    try:
        results=r.html.find(sel)
        print(results)#results是列表
        for result in results:#而result是列表中的元素
            print(result)
            mytext=result.text
            mylink=list(result.absolute_links)[0]
            mylist.append((mytext,mylink))#(mytext,mylink)组成一个元组
        return mylist
    except:
        return None
print(get_text_link_from_sel(sel))    
```

```python
sel='body > div.note > div.post > div.article > div.show-content > div > p:nth-child(6) > a'
```

```python
print(get_text_link_from_sel(sel)) 
```

```python
print(get_text_link_from_sel(sel))
```
